#### Regression: Manually-selected Variables

```python
df_models_addcol = df_formodels.copy(deep=True)
```

```python
df_models_addcol['Iron2'] = df_models_addcol.Iron**2
df_models_addcol['Protein2'] = df_models_addcol.Protein**2
df_models_addcol['Iron3'] = df_models_addcol.Iron**3
df_models_addcol['Protein3'] = df_models_addcol.Protein**3
```

```python
model_lr3 = LinearRegression()
X2 = df_models_addcol[['Iron', 'Iron2', 'Iron3', 'Protein', 'Protein2', 'Protein3']]
y2 = df_models_addcol['Selenium']
model_lr3.fit(X2, y2)
model_lr3.score(X2, y2)
```

```python
sorted(zip(X2.columns,abs(model_lr3.coef_)), key=lambda x: -x[1])
```

```python
model_lr4 = LinearRegression()
X3,y3 = df_models_addcol.drop('Selenium', axis=1),df_models_addcol['Selenium']
model_lr4.fit(X3, y3)
model_lr4.score(X3, y3)
```
